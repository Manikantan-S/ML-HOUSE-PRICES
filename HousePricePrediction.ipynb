```python
# Linear Model for the  train and test
y = np.log(train.SalePrice)
X = data.drop(['SalePrice', 'Id'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
                                    X, y, random_state=42, test_size=.33)
```

```python
import numpy as np
from sklearn import ensemble


lr = ensemble.GradientBoostingRegressor()

# Manual Ridge Regression Implementation
def manual_ridge_regression(X, y, alpha):
    # Add bias term (intercept) to X
    X_b = np.c_[np.ones((X.shape[0], 1)), X]
    
    # Compute (X^T * X + alpha * I)⁻¹ * X^T * y
    I = np.eye(X_b.shape[1])  # Identity matrix
    I[0, 0] = 0  # Don't regularize the bias term (intercept)
    
    theta = np.linalg.inv(X_b.T.dot(X_b) + alpha * I).dot(X_b.T).dot(y)
    
    return theta

# Make predictions using manually computed Ridge Regression for different alpha values
def ridge_regression_predictions(X_train, y_train, X_test, alphas):
    predictions = {}
    for alpha in alphas:
        theta = manual_ridge_regression(X_train, y_train, alpha)
        X_test_b = np.c_[np.ones((X_test.shape[0], 1)), X_test]
        preds = X_test_b.dot(theta)
        predictions[alpha] = preds
    return predictions

# Specify the range of alphas for Ridge Regression
alphas = [10**i for i in range(-2, 3)]

# Get the predictions for each alpha
ridge_predictions = ridge_regression_predictions(X_train, y_train, X_test, alphas)
```

```python
# Manual Ridge Regression Implementation
def manual_ridge_regression(X, y, alpha):
    # Add bias term (intercept) to X
    X_b = np.c_[np.ones((X.shape[0], 1)), X]
    
    # Compute (X^T * X + alpha * I)⁻¹ * X^T * y
    I = np.eye(X_b.shape[1])  # Identity matrix
    I[0, 0] = 0  # Don't regularize the bias term (intercept)
    
    theta = np.linalg.inv(X_b.T.dot(X_b) + alpha * I).dot(X_b.T).dot(y)
    
    return theta

```

```python
# Make predictions using manually computed Ridge Regression for different alpha values
def ridge_regression_predictions(X_train, y_train, X_test, alphas):
    predictions = {}
    for alpha in alphas:
        theta = manual_ridge_regression(X_train, y_train, alpha)
        X_test_b = np.c_[np.ones((X_test.shape[0], 1)), X_test]
        preds = X_test_b.dot(theta)
        predictions[alpha] = preds
    return predictions
```

```python
# Specify the range of alphas for Ridge Regression
alphas = [10**i for i in range(-2, 3)]

```

```python
# Get the predictions for each alpha
ridge_predictions = ridge_regression_predictions(X_train, y_train, X_test, alphas)
```

```python
# Calculate and print R^2 and RMSE for each alpha value
for alpha, preds in ridge_predictions.items():
    r2_score = 1 - (mean_squared_error(y_test, preds) / np.var(y_test))
    rmse = mean_squared_error(y_test, preds, squared=False)  # Use squared=False for RMSE
    print(f"Alpha: {alpha}, R^2: {r2_score:.4f}, RMSE: {rmse:.4f}")
```
